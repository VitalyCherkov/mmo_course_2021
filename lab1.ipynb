```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('arabica_data_cleaned.csv', index_col=0)
```

```python
data.shape
```

```python
data.dtypes
```

```python
data.head()
```

# Корреляция вкусовых характеристик между собой

```python
taste_cols = ['Aroma', 'Flavor', 'Aftertaste', 'Acidity', 'Body', 'Balance']
```

```python
plt.figure(figsize=(13,10))
sns.heatmap(data[taste_cols].corr(), cmap="Oranges", annot=True, linewidths=3)
```

# Корреляция вкусовых характеристик с финальным счётом

```python
final_taste_cols = taste_cols + ['Total.Cup.Points']
```

```python
plt.figure(figsize=(13,10))
sns.heatmap(data[final_taste_cols].corr(), cmap="Oranges", annot=True, linewidths=3)
```

```python
plt.figure(figsize=(13,10))
sns.pairplot(data[final_taste_cols])
```

# Зависимость финальных результатов от региона и стран участников

```python
plt.figure(figsize=(10,50))
sns.set_style("whitegrid")
ax = sns.boxplot(x="Total.Cup.Points", y="Region", data=data, palette="Set3")
```

```python
plt.figure(figsize=(10,50))
sns.set_style("whitegrid")
ax = sns.boxplot(x="Total.Cup.Points", y="Country.of.Origin", data=data, palette="Set3")
```

```python
plt.figure(figsize=(10,50))
sns.set_style("whitegrid")
ax = sns.boxplot(x="Total.Cup.Points", y="Processing.Method", data=data, palette="Set3")
```

```python
sorted_data = data.sort_values(by=['Total.Cup.Points'], ascending=False)
top_200 = sorted_data[:200]

plt.figure(figsize=(10,50))
sns.set_style("whitegrid")
ax = sns.violinplot(x="Total.Cup.Points", y="Region", data=top_200, palette="Set3")
```

# Распределение по странам-участникам соревнования

```python
plt.figure(figsize=(15,15))
data.groupby('Country.of.Origin').size().plot(kind='pie')
```

# Посмотрим, как зависит результат от производителя

```python
plt.figure(figsize=(20,5))

sns.stripplot(x="Producer", y="Total.Cup.Points", data=data);
```

```python
plt.figure(figsize=(20,50))

sns.stripplot(y="Producer", x="Total.Cup.Points", data=top_200);
```
