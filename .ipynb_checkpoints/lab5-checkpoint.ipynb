```python
from spacy.lang.ru import Russian
import spacy
from spacy import displacy
```

```python
!spacy download ru_core_news_sm
```

```python
sentence = """
В ответ на официальный запрос мы удалили некоторые результаты с этой страницы. Вы можете ознакомиться с запросом на сайте В ответ на официальный запрос мы удалили некоторые результаты с этой страницы. Вы можете ознакомиться с запросом на сайте. 
"""

nlp = spacy.load('ru_core_news_sm')
text = nlp(sentence)
text
```

```python
for token in text:
    print('{} - {} - {}'.format(token.text, token.pos_, token.dep_))
```

```python
# лемматизация
for token in text:
      print(token, token.lemma, token.lemma_)
```

```python
# выделение (распознавание) именованных сущностей
for ent in text.ents:
    print(ent.text, ent.label_)
```

```python
displacy.render(text, style='ent', jupyter=True)
```

```python
displacy.render(text, style='dep', jupyter=True)
```
